```python
from elitho.mask_pattern import LinePattern
from elitho import const
mask = LinePattern(cd=80, gap=60, direction="V", field_type="DF")(const.NDIVX, const.NDIVY)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(mask, cmap='gray', interpolation='nearest')
plt.title("Mask")
plt.colorbar(label="Pixel value")
plt.show()
```

```python
from elitho import intensity
intensity_map = intensity.intensity(mask, polar="X", is_high_na=True)
```

```python
from matplotlib import pyplot as plt
plt.imshow(intensity_map)
plt.title("Intensity")
plt.colorbar(label="Pixel value")
plt.show()
```
